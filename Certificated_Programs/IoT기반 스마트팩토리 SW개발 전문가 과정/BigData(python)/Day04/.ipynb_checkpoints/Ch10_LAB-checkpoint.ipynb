#### 연설문 데이터 분석

```python
word = dict()
with open('input.txt','r') as a:
    line = a.readline()
    while(line != ""):
        words = line.split()
        for i in words:
            item = i.strip(' ,.:;-')
            item = item.lower()
            if not( item in list(word.keys()) ):
                word[item] = 1
            else:
                word[item] += 1        
        
        line= a.readline()
with open('output.txt','w') as a:
    for k,v in word.items():
        a.write("%s\t%s\n" %(k,v))
```

```python
totalList = list()
total = 0
with open('강수량.csv','r') as csv:
    line = csv.readline()
    line = line.replace('\n','')
    words = line.split(',')
    line = csv.readline()    
    line = line.replace('\n','')
    while line!="":
        
        line = line.replace('\n','')
        temp = line.split(',')
        di = dict()
        di[words[0]] = temp[0]
        di[words[1]] = int(temp[1])
        di[words[2]] = float(temp[2])
        total+=float(temp[2])
        totalList.append(di)
        
        line = csv.readline()
totalavg = total/len(totalList)
        
print('서울 2009년 1월부터 21년 5월까지 총 강수량 : ',total)
print('서울 2009년 1월부터 21년 5월까지 평균 강수량 : ',totalavg)
        
        
```

```python

def getWord():
    import random
    with open('hangman.txt') as fr:
        lines = fr.readlines()
        word = lines[random.randrange(0,len(lines))].strip()
    return word

def hideWord(word):
    outputWord = []
    for i in range(0,len(word)):
        outputWord += '_'
    return outputWord

def replace(output, cha ,word):
    for i in range(0,len(word)):
        if word[i] == cha :
            output[i] = cha
    return output


word = getWord()
#print(word)
outputWord = hideWord(word)
print(outputWord)
life = 10
done = True

while done:
    inputChar = input("단어를 추측하세요 : ")
    inputChar = inputChar.strip()[0]
    if inputChar == "":
        continue
    elif inputChar in word :
        outputWord = replace(outputWord,inputChar,word)
        if not('_' in outputWord):
            done = False        
    else:
        life-=1
        if life <= 0 :
            done = False
    print(outputWord)
else:
    print("성공입니다")
```
