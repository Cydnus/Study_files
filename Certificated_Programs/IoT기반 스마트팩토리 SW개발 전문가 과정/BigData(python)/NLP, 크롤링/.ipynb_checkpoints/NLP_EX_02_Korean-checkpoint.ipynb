````python
import json
import re
from konlpy.tag import Okt
from collections import Counter
import matplot
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
from wordcloud import WordCloud
````

데이터로드

````python
inputFile_Name = './etnews.kr_facebook_2016-01-01_2018-08-01_4차 산업혁명'
data = json.loads(open(inputFile_Name+'.json','r',encoding='utf-8').read())
data
````

문자나 숫자가 아닌거 제외

````python
message = ''
for item in data:
    if 'message' in item.keys():
        message = message + re.sub(r'[^\w]',' ',item['message']) +""
message
````

명사만 추출

````python
nlp = Okt()
message_N = nlp.nouns(message)
message_N
````

````python
count = Counter(message_N)
count
````

가장 많이 나온 상위 80개 추출  

````python
word_count = dict()
for tag,counts in count.most_common(80):
    if( len(str(tag))>1):
        word_count[tag] = counts
        print('%s : %d' %(tag,counts))
````

폰트설정

````python
font_path ='c:/Users/kccistc/Downloads/font/Maplestory OTF Light.otf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
````

````python
plt.figure(figsize=(12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')

plt.grid(True)
sorted_key = sorted(word_count,key=word_count.get,reverse=True)
sorted_value = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)),sorted_value,align='center')
plt.xticks(range(len(word_count)), list(sorted_key),rotation='75')
plt.show()
````

````python
wc = WordCloud(font_path, background_color='ivory', width = 800 , height=600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
````

````python
cloud.to_file(inputFile_Name +'_cloud.jpg')
````
