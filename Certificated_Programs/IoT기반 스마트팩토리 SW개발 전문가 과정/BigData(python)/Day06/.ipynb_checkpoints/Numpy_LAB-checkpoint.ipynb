BMI 계산하기

```python
import numpy as np

heights = [1.83,1.76,1.69,1.86,1.77,1.73]
weights = [86,74,59,95,80,68]

h = np.array(heights)
w = np.array(weights)

result = w /(h**2)

print(result)
```

잡음이 들어간 직선 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

pure = np.linspace(0,10,100)           # 0~10사이 100개
noise = np.random.normal(0,1,100)      # 평균이 0이고 표준편차가 1인 100개의 난수

signal = pure + noise

plt.plot(signal)
plt.show()


```
