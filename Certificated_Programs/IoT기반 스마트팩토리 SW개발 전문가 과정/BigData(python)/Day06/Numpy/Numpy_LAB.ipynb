BMI 계산하기

```python
import numpy as np

heights = [1.83,1.76,1.69,1.86,1.77,1.73]
weights = [86,74,59,95,80,68]

h = np.array(heights)
w = np.array(weights)

result = w /(h**2)

print(result)
```

잡음이 들어간 직선 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

pure = np.linspace(0,10,100)           # 0~10사이 100개
noise = np.random.normal(0,1,100)      # 평균이 0이고 표준편차가 1인 100개의 난수

signal = pure + noise

plt.plot(signal)
plt.show()


```

두 개의 정규 분포를 그리기

```python
import numpy as np 
import matplotlib.pyplot as plt

m,sigma = 10,2
a = np.random.randn(10000)
b = m + sigma *np.random.randn(10000)

plt.figure(figsize=(10,6))
plt.hist(a,bins=20)
plt.hist(b,bins=20)
plt.show()
```

사인함수 그리기

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi,2*np.pi)
y = np.sin(x)

y1 = np.sin(x) *3

plt.plot(x,y)
plt.plot(x,y1)
plt.show()


```

MSE오차 계산하기

```python
import numpy as np 

ypred = np.array([1,0,0,0,0])
y = np.array([0,1,0,0,0])
n = 5
mse=(1/n)*np.sum( np.square(ypred-y ) )
print(mse)
```

그래프 그리기
    
linspace()로 x축값을 생성하고, $f(x)=1,f(x)=x,f(x)=x^2$ 의 그래프를 그려보자

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

y1 = np.ones(100)+1
y2 = x
y3 = x**2

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```
