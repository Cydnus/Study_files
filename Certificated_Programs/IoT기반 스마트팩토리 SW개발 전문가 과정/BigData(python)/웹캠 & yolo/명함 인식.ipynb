```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Edge Detecton

```python
image = cv2.imread('document3.jpg')

orig = image.copy()
```

```python
r = 800.0/image.shape[0]
dim = (int(image.shape[1] * r) ,800)

image = cv2.resize(image,dim,interpolation = cv2.INTER_AREA)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray,(3,3),0)
edged = cv2.Canny(gray,75,200)

```

```python
plt.figure(figsize=(20,10))

plt.suptitle('STEP 1 : Edge Detection')
plt.subplot(1,2,1)
plt.imshow(image)
plt.axis(False)
plt.title('image')

plt.subplot(1,2,2)
plt.imshow(edged, cmap='gray')
plt.title('Edged')
plt.axis(False)
plt.show()
```

```python
screenCnt = np.array([])

ret,thresh = cv2.threshold(edged.copy(),127,255,0)

temp = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnts = temp[0] if len(temp) == 2 else temp[1]

cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:5]

for c in cnts:
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c , 0.02 * peri, True)
    if len(approx) == 4:
        screenCnt = approx
        break
```

```python
cv2.drawContours(image, [screenCnt], -1, (0,255,0), 2)

plt.figure(figsize=(10,10))
plt.title('STEP 2 : Find Contours Of Paper')
plt.imshow(image)
plt.axis(False)
plt.show()
```

```python
def order_points(pts):
    rect = np.zeros((4,2),dtype ="float32")
    s = pts.sum(axis = 1)
    
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    
    diff = np.diff(pts,axis = 1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    
    return rect
```

```python
rect = order_points(screenCnt.reshape(4,2)/r)
(tl,tr,br,bl) = rect 

w1 = abs(br[0]-bl[0])
w2 = abs(tr[0]-tl[0])
h1 = abs(tr[1]-br[1])
h2 = abs(tl[1]-bl[1])

maxwidth = max([w1,w2])
maxheight = max([h1,h2])

dst = np.float32([[0,0],[maxwidth-1,0],[maxwidth-1,maxheight-1],[0,maxheight-1]])

M = cv2.getPerspectiveTransform(rect, dst)
img = orig.copy()


warped = cv2.warpPerspective(img, M, (int(maxwidth), int(maxheight))) 
```

```python
plt.figure(figsize=(10,10))
plt.title('STEP 3 : Apply perspective transform')
plt.imshow(warped)
plt.axis(False)
plt.show()
```

```python
warped = cv2.cvtColor(warped, cv2.COLOR_BGR2GRAY)
warped = cv2.adaptiveThreshold(warped, 
                               255, 
                               cv2.ADAPTIVE_THRESH_MEAN_C, 
                               cv2.THRESH_BINARY,
                               21,
                               10  )
```

```python
fig = plt.figure(figsize=(20,10))
plt.suptitle('STEP 4 : Apply Adaptive Threshold')
plt.subplot(1,2,1)
plt.title('image')
plt.axis(False)
plt.imshow(orig)
plt.subplot(1,2,2)
plt.imshow(edged, cmap='gray')
plt.title('Edged')
plt.axis(False)
plt.show()

fig = plt.figure(figsize=(20,10))
plt.imshow(warped)
plt.title('warped')

plt.axis(False)

plt.show()
```
