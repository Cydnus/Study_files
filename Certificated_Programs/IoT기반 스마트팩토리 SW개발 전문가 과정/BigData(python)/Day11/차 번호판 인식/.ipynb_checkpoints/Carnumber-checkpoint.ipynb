```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
```

```python
def ImageRead(imagePath):
    img = cv2.imread(imagePath)
    return img, img.shape
```

```python
def PrintImage(image):
    plt.figure(figsize = (12,10))
    plt.imshow(image, cmap='gray')
    plt.axis(False)
    plt.show()
```

```python
def ToGray(image) :
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return gray
```

```python
def ToBluringAndThreshold(gray_image):
    b = cv2.GaussianBlur(gray_image, ksize=(5,5),sigmaX=0)
    
    bt = cv2.adaptiveThreshold(
        b,
        maxValue = 255.0,
        adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        thresholdType=cv2.THRESH_BINARY_INV,
        blockSize=19,
        C=9
    )
    return bt
    
```

```python
def FindContours(image, shape):
    con = cv2.findContours( image,
                           mode=cv2.RETR_LIST,
                           method=cv2.CHAIN_APPROX_SIMPLE)
    contours = con[0] if len(con)==2 else con[1]
    result= np.zeros(shape,dtype=np.uint8)
    
    result = cv2.drawContours(result,
                              contours=contours,
                              contourIdx=-1,
                              color=(255,255,255))
    return result, contours
```

```python
def PrepareData(image, contours, shpae):
    result = np.zeros(shape,dtype=np.uint8)
    contours_dict = []
    
    for contour in contours:
        x,y,w,h = cv2.boundingRect(contour)
        cv2.rectangle(result, 
                      pt1 = (x,y), 
                      pt2=(x+w,y+h),
                      color=(255,255,255),
                      thickness = 2
                     )
        contours_dict.append(
        {
            'contour': contour,
            'x' : x,
            'y' : y,
            'w' : w,
            'h' : h,
            'cx' : x+(w/2),
            'cy' : y+(h/2)
        })
        
    return result, contours_dict
        
    
```

```python
def selectCandidate_CharSize (image, contours_dict, shape):
    MIN_AREA, MIN_WIDTH, MIN_HEIGHT = 80, 2, 8
    MIN_RATIO, MAX_RATIO = 0.25,1
    
    possible_contour = []
    cnt = 0
    for d in contours_dict:
        area = d['w'] * d['h']
        ratio = d['w'] / d['h']
        
        if (area > MIN_AREA and
            d['w'] > MIN_WIDTH and
            d['h'] > MIN_HEIGHT and 
            MIN_RATIO < ratio < MAX_RATIO):
            d['idx'] = cnt
            cnt += 1
            possible_contour.append(d)
            
    temp_result = np.zeros(shape, dtype=np.uint8)
    for d in possible_contour:
        cv2.rectangle(temp_result, 
                     pt1=(d['x'],d['y']),
                     pt2=(d['x']+d['w'],d['y']+d['h']),
                      color = (255,255,255),
                      thickness = 2
                     )
    return temp_result, possible_contour
    
        
        
```

```python
def selectCandidate_ArrangementContours(listContour, share):
    MAX_DIAG_MULTIPLYER = 5
    MAX_ANGLE_DIFF = 12
    MAX_AREA_DIFF= 0.5
    MAX_WIDTH_DIFF = 0.8
    MAX_HEIGHT_DIFF = 0.2
    MIN_N_MATCHED = 3
    
    def find_chars(contor_list):
        result_idx = []
        for d1 in contor_list :
            contours_idx =[]
            for de in contour_list:
                if d1['idx'] == d2['idx'] : continue
                
                dx = abs(d1['cx'] - d2['cx'])
                dy = abs(d1['cy'] - d2['cy'])
                
                diagonal_length = np.sqrt(d1['w'] ** 2+d1['h'] ** 2 )
                distance = np.linalg.norm(
                    np.array([dx1['cx'],d1['cy']]) 
                    - np.array([dx2['cx'] , d2['cy']])
                )
                angle_diff = 90 if dx == 0 else np.degress(np.arctan(dy/dx))
                area_diff = abs(d1['w'] * d1['h'] - d2['w'] * d2['h']) / (d1['w'] *d1['h'])
                width_diff = abs(d1['w'] - d2['h']) / d1['h']
                
                if (distance < diagonal_length * MAX_DIAG_MULTIPLYER and
                   angle_diff < MAX_ANGLE_DIFF and
                   area_diff < MAX_AREA_DIFF and
                    width_diff < MAX_WIDTH_DIFF and
                    height_diff < MAX_HEIGHT_DIFF  ) :
                    contours_idx.append(d2['idx'])
                
            contours_idx.append(d1['idx'])    
            
            if len(contours_idx) < MIN_N_MATCHED:
                continue

            result_idx.append(contours_idx)

            unmatched_idx = []
            for d4 in contor_list:
                if d4 ['idx'] not in contours_idx:
                    unmatched_idx.append(d4['idx'])

            unmatched_contour = np.take(listContour, unmatched_idx)
            recursive_contour_list = find_chars(unmatched_contour)
            for idx in recursive_contour_list:
                result_idx.append(idx)

            break
        return result_idx
    result = find_chars(listContour)
    matched_result = []
    for ind_list in result:
        matched_result.append(np.take(listContour, ind_list))

    temp_result = np.zeros(shape, dtype=np.uint8)

    for r in matched_result:
        for d in r:
            cv2.rectangle(temp_result, 
                         pt1=(d['x'],d['y']),
                         pt2=(d['x']+d['w'],d['y']+d['h']),
                          color = (255,255,255),
                          thickness = 2
                         )
    return temp_result, matched_result
```

```python
if __name__ == "__main__":
    plt.style.use('dark_background')
    
    img_ori,shape =ImageRead('image/01.jpg')
    PrintImage(img_ori)
    
    gray = ToGray(img_ori)
    PrintImage(gray)
    
    threshold = ToBluringAndThreshold(gray)
    PrintImage(threshold)
    
    contour_img, contours = FindContours(threshold, shape)
    PrintImage(contour_img)
    
    prepare_img, contours_dict = PrepareData(contour_img,  contours, shape )
    PrintImage(prepare_img)
    
    candidateimage, cantidates = selectCandidate_CharSize(prepare_img,contours_dict, shape )
    PrintImage(candidateimage)
    
    scacImg, candidates = selectCandidate_ArrangementContours(contours_dict, shape)
    PrintImage(scacImg)
```
